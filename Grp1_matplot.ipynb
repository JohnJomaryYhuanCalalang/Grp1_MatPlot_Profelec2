```python
import matplotlib.pyplot as plt
import numpy as np

agents = ['Phoenix', 'Sova', 'Jett', 'Brimstone']
pick_rates = [25, 30, 20, 25]
win_rates = [40, 25, 20, 15]

# Plotting
plt.figure(figsize=(12, 5))

# Subplot 1
plt.subplot(1, 2, 1)
plt.bar(agents, pick_rates, color='skyblue')
plt.title('Agent Pick Rate')
plt.xlabel('Agents')
plt.ylabel('Pick Rate')
plt.grid(axis='y')

# Subplot 2
plt.subplot(1, 2, 2)
plt.pie(win_rates, labels=agents, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightgreen', 'lightskyblue', 'lightyellow'])
plt.title('Agent Win Rate')

plt.tight_layout()
plt.show()
```
